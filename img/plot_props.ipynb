```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
L = [16,32,64,128]

for iL in L :
    path = f'./../data/outvars_L_{iL}.dat'
    data = np.loadtxt(path)
    plt.plot(data[:,1],data[:,2])

plt.xlim(.09,0.15)

plt.xlabel('$p$',fontsize=20,rotation=0)
plt.ylabel('$M$',fontsize=20,rotation=0)
plt.legend(loc='best',fontsize=16)
plt.savefig('magnet.png')
```
